```python
import sqlite3
import pandas as pd
import os
import time
import psutil
```

```python
# -- ĐÁNH GÁI HIỆU NĂNG HỆ THỐNG ---

print("=" * 60)
print("ĐÁNH GIÁ HIỆU NĂNG HỆ THỐNG")
print("=" * 60)

db_path = "database/traffic_sql.db"
csv_path = "traffic_test_output_01.csv"

# ==========================================
# 1. TỐC ĐỘ ĐỌC DỮ LIỆU
# ==========================================

print("\n--- TỐC ĐỘ ĐỌC DỮ LIỆU ---")

start = time.time()

conn = sqlite3.connect(db_path)
df_sql = pd.read_sql("SELECT * FROM traffic_data", conn)

sqlite_time = time.time() - start

start = time.time()
df_csv = pd.read_csv(csv_path)
csv_time = time.time() - start

print(f"Thời gian đọc SQLite: {sqlite_time:.4f} giây")
print(f"Thời gian đọc CSV:    {csv_time:.4f} giây")

# ==========================================
# 2. THỜI GIAN XỬ LÝ
# ==========================================

print("\n--- THỜI GIAN XỬ LÝ ---")

n_records = len(df_sql)

print(f"Tổng số bản ghi: {n_records}")

if n_records > 0:
    avg_time = sqlite_time / n_records
else:
    avg_time = 0

print(f"Thời gian trung bình / bản ghi: {avg_time:.8f} giây")

# ==========================================
# 3. CPU + RAM
# ==========================================

print("\n--- TÀI NGUYÊN HỆ THỐNG ---")

cpu_usage = psutil.cpu_percent(interval=1)
memory_usage = psutil.Process(os.getpid()).memory_info().rss / 1024 / 1024

print(f"CPU sử dụng: {cpu_usage:.2f}%")
print(f"RAM sử dụng: {memory_usage:.2f} MB")

# ==========================================
# 4. CHI PHÍ LƯU TRỮ
# ==========================================

print("\n--- CHI PHÍ LƯU TRỮ ---")

db_size = os.path.getsize(db_path) / 1024 / 1024
csv_size = os.path.getsize(csv_path) / 1024 / 1024

print(f"Kích thước SQLite: {db_size:.4f} MB")
print(f"Kích thước CSV:    {csv_size:.4f} MB")

# ==========================================
# 5. TỐC ĐỘ TRUY VẤN
# ==========================================

print("\n--- TỐC ĐỘ TRUY VẤN ---")

start = time.time()

query = """
SELECT *
FROM traffic_data
WHERE gaussian_hmm_state = 1
"""

result = pd.read_sql(query, conn)

query_time = time.time() - start

print(f"Thời gian query state=1: {query_time:.6f} giây")
print(f"Số dòng trả về: {len(result)}")

conn.close()

print("\n" + "=" * 60)
print("KẾT THÚC")
print("=" * 60)
```
